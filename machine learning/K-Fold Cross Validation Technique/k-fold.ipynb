# What is K-Fold Cross Validation?

K-fold cross-validation is a technique for evaluating predictive models. The dataset is divided into k subsets or folds. The model is trained and evaluated k times, using a different fold as the validation set each time. Performance metrics from each fold are averaged to estimate the model’s generalization performance. This method aids in model assessment, selection, and hyperparameter tuning, providing a more reliable measure of a model’s effectiveness.

In each set (fold) training and the test would be performed precisely once during this entire process. It helps us to avoid overfitting. As we know when a model is trained using all of the data in a single short and give the best performance accuracy. To resist this k-fold cross-validation helps us to build the model is a generalized one.

To achieve this K-Fold Cross Validation, we have to split the data set into three sets, Training, Testing, and Validation, with the challenge of the volume of the data.

Here Test and Train data set will support building model and hyperparameter assessments.

In which the model has been validated multiple times based on the value assigned as a parameter and which is called K and it should be an INTEGER.

Make it simple, based on the K value, the data set would be divided, and train/testing will be conducted in a sequence way equal to K time.

# K-Fold cross-validation helps prevent overfitting through several methods:

1. **Repeated model evaluation**:
The model is trained and evaluated in K different stages, each time using a different portion of the data as the test set. This ensures that the model tests its performance on all parts of the data, not just relying on a specific subset.

2. **Utilization of all data for training and testing**:
With K-Fold Cross Validation, all data is used for both training and testing. This helps the model gain a better understanding of the entire dataset and prevents over-reliance on a particular subset of the data.

3. **Performance comparison**:
Performance metrics of the model are calculated in each fold and then averaged. This average performance gives us a comprehensive picture of the model's performance, reducing the likelihood that the model performs well only on a specific portion of the data.

4. **Better hyperparameter tuning**:
Since K-Fold cross-validation is used for hyperparameter tuning, we can find optimal settings using more extensive data. This helps prevent the creation of models that only perform well on a specific subset of the data.

# Thumb Rules Associated with K Fold

* Now, we will discuss a few thumb rules while playing with K – fold :


    1. **When K = 2**:
       In this case, there are just two iterations. In each iteration, one data point is used for testing, and the remaining data point is used for training. Essentially, it's a simple holdout validation where the dataset is split into two parts: one for training and the other for testing. This is not technically considered Leave-One-Out Cross-Validation (LOOCV), as LOOCV implies using all but one data point for training and leaving out just one for testing.
    
    2. **When K = Number of Records in the Dataset**:
       When K is set equal to the number of records in the dataset (n), it indeed becomes Leave-One-Out Cross-Validation (LOOCV). In this scenario, there is one iteration per data point. In each iteration, one data point is left out as the test set, and the remaining n-1 data points are used for training the model. This process is repeated n times, with each data point being left out once for testing. 
    
       The main advantage of LOOCV is that it provides an unbiased estimate of the model's performance on unseen data since every data point is used for testing exactly once. However, it can be computationally expensive, especially for large datasets, as it requires fitting the model n times.


* The optimized value for the K is 10 and used with the data of good size. (Commonly used)

* If the K value is too large, then this will lead to less variance across the training set and limit the model currency difference across the iterations.

    When the value of K (the number of folds) in cross-validation is too large, it can lead to less variance across the training sets and limit the model's ability to capture differences across the iterations. Here's why:
    
    1. **Less Variance Across Training Sets**:
       With a larger value of K, each fold will contain a larger portion of the total dataset. As a result, the training sets in each fold will be more similar to each other. This reduces the variability across training sets because each training set overlaps significantly with the others. As a consequence, the model may not be exposed to enough diverse patterns and variations present in the data, potentially leading to a less robust model.
    
    2. **Limited Model Currency Difference Across Iterations**:
       When the training sets are highly similar due to a large value of K, the model's performance across iterations may not vary significantly. This is because the training sets have similar characteristics, and the model is trained on almost the same data in each iteration. As a result, the differences in model performance across iterations become less pronounced, limiting the ability to detect variations in model behavior and potentially hiding issues such as overfitting or underfitting.
            


* The number of folds is indirectly proportional to the size of the data set, which means, if the dataset size is too small, the number of folds can increase.

* Larger values of K eventually increase the running time of the cross-validation process.


Please remember K-Fold Cross Validation for the below purpose in the ML stream.

    Feature selection





# code with python

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = load_iris()
x = data.data
y = data.target

model = LogisticRegression(max_iter=400)

k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

accuracies = list()

for train_index, test_index in kf.split(x):
    x_test , y_test = x[test_index] , y[test_index]
    x_train, y_train = x[train_index] , y[train_index]
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_pred, y_test)
    accuracies.append(accuracy)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))

x = list(range(1, k + 1))
y = accuracies
x = list(map(lambda x : str(x), x))

for fold, accuracy in zip(x, y):
    plt.text(fold, accuracy, f'{accuracy:.2f}', fontsize=9, ha='center', va='bottom')
plt.plot(x, y, marker='o', linestyle='--')
plt.xlabel('number of k train')
plt.ylabel('accuracy')
plt.title('accuracies for each train')
plt.grid(True)
plt.show()
```
